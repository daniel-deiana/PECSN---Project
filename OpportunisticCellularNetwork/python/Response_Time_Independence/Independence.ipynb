```python
import numpy as np
import pandas as pd
import matplotlib as mpl 
import scipy.stats as st
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from statsmodels.graphics.tsaplots import plot_acf

rate = [0.25,0.50,1,2.50,3.75,5,6.25]
rate_ = ['0_25','0_50','1_00','2_50','3_75','5_00','6_25']

for i in range(0,len(rate)):
    
    raw = pd.read_csv('independence' + rate_[i] + '.csv')
    
    plot_acf(raw['value'], zero=False, auto_ylims=True, lags=100, alpha=.05)
    plt.xlabel('lag')
    plt.title('Autocorrelation - $\lambda$ = ' + str(rate[i]))
    plt.savefig('./plots/Autocorrelation_lambda' + str(rate[i]) + '.png', bbox_inches='tight')
    plt.show()
    
    plt.scatter(raw['value'][0:len(raw['value'])-2],raw['value'][1:len(raw['value'])-1],s=1)
    plt.title('LAG = 1 - $\lambda$ = ' + str(rate[i]))
    plt.ylabel('$X_{i + LAG}$')
    plt.xlabel('$X_i$')
    plt.savefig('./plots/LagPlot_lambda' + str(rate[i]) + '.png', bbox_inches='tight')
    plt.show()
```

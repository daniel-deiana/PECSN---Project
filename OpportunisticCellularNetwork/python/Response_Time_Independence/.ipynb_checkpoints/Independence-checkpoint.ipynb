```python
import numpy as np
import pandas as pd
import matplotlib as mpl 
import scipy.stats as st
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

correlation = []

raw = pd.read_csv('independence.csv')

mean = np.mean(raw['value'])
s2   = np.std(raw['value'])*np.std(raw['value'])

length = 1000

for i in range(1,length):
    tmp = 0
    for j in range(0,length-i):
        tmp += ((raw['value'][j]-mean)*(raw['value'][j+i]-mean))
    tmp /= s2
    tmp /= len(raw['value'])
    correlation.append(tmp)
    
plt.scatter(range(1,len(correlation)+1),correlation,s=1)
plt.show()
    
```

```python
import numpy as np
import pandas as pd
import matplotlib as mpl 
import scipy.stats as st
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

raw = pd.read_csv('independence.csv')
    
plt.scatter(raw['value'][0:len(raw['value'])-2],raw['value'][1:len(raw['value'])-1],s=1)
plt.title('LAG = 1')
plt.ylabel('$X_{i + LAG}$')
plt.xlabel('$X_i$')
plt.show()

plt.scatter(raw['value'][0:len(raw['value'])-3],raw['value'][2:len(raw['value'])-1],s=1)
plt.title('LAG = 2')
plt.ylabel('$X_{i + LAG}$')
plt.xlabel('$X_{i}$')
plt.show()

plt.scatter(raw['value'][0:len(raw['value'])-101],raw['value'][100:len(raw['value'])-1],s=0.5)
plt.title('LAG = 100')
plt.ylabel('$X_{i + LAG}$')
plt.xlabel('$X_{i}$')
plt.show()

plt.scatter(raw['value'][0:len(raw['value'])-1001],raw['value'][1000:len(raw['value'])-1],s=0.5)
plt.title('LAG = 1000')
plt.ylabel('$X_{i + LAG}$')
plt.xlabel('$X_{i}$')
plt.show()
```
